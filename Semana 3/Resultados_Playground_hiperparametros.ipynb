## Respuestas a las preguntas planteadas

### ¿Es buena o mala la configuración? ¿Por qué?
- **Buena:** Training loss y Test loss **bajos y parecidos** → la red aprende y generaliza. Si además converge rápido (LR adecuado), mejor.
- **Mala:** (1) Loss que no baja o **oscila** → LR demasiado alto. (2) **Train muy bajo y test alto** → overfitting (modelo demasiado complejo o falta regularización). (3) Loss que **baja muy lento** → LR demasiado bajo. (4) En problemas no lineales (círculo, XOR, curva), **falta capacidad** (más neuronas/capas) o **features** (X1², X1·X2, etc.).

**Por qué cambiar los datos mejora o empeora (Ej. 1 – Two clusters)**  
El problema es **linealmente separable**: una sola recta basta para separar los dos grupos. Por eso 1 neurona alcanza.

- **Learning rate bajo (0.01):** Los pesos se actualizan poco a poco → el loss baja de forma estable pero más lenta. Mejor que subir mucho el LR sin necesidad.

- **ReLU vs Tanh:** ReLU suele bajar el loss más rápido y dar un mínimo ligeramente mejor aquí; Tanh también funciona. Cambiar la activación puede **mejorar** la velocidad de aprendizaje y a veces el error final.

**Por qué cambiar los datos mejora o empeora (Ej. 3 – Gaussian / overfitting)**  
El objetivo es aprender la frontera **sin memorizar** los puntos. Si la red es demasiado grande, ajusta hasta el ruido y falla en test.

- **Modelo simple (1 capa, 2 neuronas):** Tiene poca capacidad → no puede encajar todos los detalles de los datos. El training loss no baja tanto, pero el **test loss se acerca al train**. - **Regularization > 0:** Penaliza pesos grandes y obliga a la red a usar soluciones más suaves. Subir la regularización **mejora** el test loss cuando hay overfitting (train bajo, test alto), porque reduce el sobreajuste.
- Resumen: cambiar a **menos capacidad** o **más regularización** suele mejorar el resultado en test cuando hay overfitting.

**Por qué cambiar los datos mejora o empeora (Ej. 4 – XOR)**  
XOR no es linealmente separable: con solo X1 y X2 no existe una recta que separe las clases. La red necesita **información extra** (features derivadas) para poder separar bien.

- **Solo X1, X2:** La red solo ve las coordenadas originales. No puede dibujar la frontera en forma de “X” que necesita XOR → el loss se queda alto. **Añadir features** (X1², X2², X1·X2) **mejora** porque da la información necesaria para que una capa oculta pueda construir la frontera correcta.

- **Sin X1·X2 (solo X1², X2²):** En muchos casos la red puede aproximar XOR solo con los cuadrados, pero suele ser algo peor que con X1·X2. Incluir **X1·X2 mejora** la expresividad y suele dar un error menor.

**Por qué cambiar los datos mejora o empeora (Regresión – Ej. 6 a 10)**  
En regresión se busca aproximar una función (línea o curva). Los mismos conceptos aplican: learning rate, capacidad del modelo y features.

- **Learning rate bajo (0.0001):** Los pesos cambian muy poco en cada paso → el loss baja muy lento pero estable. **Subir un poco el LR (p. ej. 0.03) mejora** el tiempo de entrenamiento sin perder estabilidad.
- **Learning rate alto (0.3, 1):** Los pasos son tan grandes que el optimizador no se acerca al mínimo y el loss oscila o no baja. **Bajar el LR mejora** porque permite converger.
- **0 capas ocultas:** La red es una regresión lineal (una recta o un plano). **Mejora** cuando la relación verdadera es lineal; si es curva, el error se queda alto.
- **Features (X1, X1², etc.):** Si la relación es no lineal (curva), hace falta X² (o más features) para que la red pueda ajustarla. **Añadir X1² (y similares) mejora** el loss cuando la función objetivo no es lineal.
- **Muchas capas/neuronas (3 capas, 8/8/8):** La red puede ajustar muy bien el entrenamiento pero **test empeora** (overfitting). **Reducir tamaño o añadir regularización mejora** el error en test.
- Resumen: ajustar **LR** para que converja, **features** para que la función sea modelable, y **tamaño del modelo** (y regularización) para que generalice bien en test.
