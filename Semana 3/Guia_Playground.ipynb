En este cuaderno te voy a guiar **paso a paso** para que puedas:

- Relacionar cada **problema del PDF** con una **configuración concreta** en `playground.tensorflow.org`.
- Saber **qué mover**, **qué observar** y **cómo interpretar** lo que pasa.
- Tener frases técnicas pero claras para explicar tus resultados.

> NOTA: Este cuaderno no entrena redes con código, sino que te sirve como **guión de trabajo** para que juegues en el Playground y luego puedas escribir tu informe.

### 0. Recordatorio rápido: partes del TensorFlow Playground

Abre este enlace de ejemplo (círculo):

`https://playground.tensorflow.org/#activation=tanh&batchSize=10&dataset=circle&regDataset=reg-plane&learningRate=0.03&regularizationRate=0&noise=0&networkShape=4,2&seed=0.82868&showTestData=false&discretize=false&percTrainData=50&x=true&y=true&xTimesY=false&xSquared=false&ySquared=false&cosX=false&sinX=false&cosY=false&sinY=false&collectStats=false&problem=classification&initZero=false&hideText=false`

En la pantalla verás:

- **Dataset (Data)**: puntos naranjas y azules que quieres separar o ajustar.
- **Network**: la red (capas ocultas, neuronas, activaciones, pesos).
- **Controls**:
  - **Epoch**: cuántas veces ha pasado por los datos.
  - **Learning rate**: tasa de aprendizaje.
  - **Activation**: función de activación de las capas ocultas (ReLU, Tanh, Sigmoid, Linear).
  - **Batch size**: tamaño de lote.
  - **Regularization** y **Regularization rate**: para controlar overfitting.
- **Features** (x, y, x², y², x·y, etc.): variables de entrada que activas o desactivas.

Idea mental importante:

- La red aprende a **colorear el fondo** (azul/naranja) para separar los puntos.
- Las **capas ocultas** generan representaciones intermedias más complejas.
- El **learning rate** controla qué tan grandes son los pasos cuando actualiza los pesos.

En las siguientes secciones, seguirás instrucciones para cada grupo de ejercicios del PDF.


### 0.1. Resumen rápido: qué pasa cuando cambio cada cosa

Esta sección es para estudiar **rápido** y saber responder preguntas tipo "¿qué pasa si cambio X?".

- **Si cambio el número de neuronas (en una capa):**
  - Más neuronas ⇒ **más capacidad** para aprender formas complejas.
  - Efecto visual: la frontera/fondo se vuelve más flexible y detallado.
  - Riesgo: si son demasiadas en un problema simple, aparece **overfitting** (frontera muy rara, pegada a cada punto).

- **Si agrego o quito capas ocultas:**
  - Más capas ⇒ la red es **más profunda**, puede componer funciones más complicadas.
  - Efecto visual: puede pasar de una línea casi recta a curvas, círculos, formas complejas.
  - Riesgo: demasiadas capas para pocos datos ⇒ overfitting + entrenamiento más inestable/lento.

- **Si cambio la función de activación (ReLU / Tanh / Sigmoid / Linear):**
  - `Linear` (lineal): la red casi solo puede hacer fronteras rectas; no aprende patrones no lineales.
  - `Sigmoid`: puede modelar no linealidades, pero **se satura** fácil ⇒ gradientes pequeños, aprende lento.
  - `Tanh`: parecida a sigmoid pero entre -1 y 1 ⇒ suele aprender mejor que sigmoid en estos ejemplos.
  - `ReLU`: 0 si x<0, x si x>0 ⇒ rápida y evita en parte el "gradiente que desaparece"; muy usada en redes profundas.

- **Si cambio el learning rate (tasa de aprendizaje):**
  - Muy bajo (0.0001) ⇒ pasos muy pequeños ⇒ el loss baja muy despacio, parece que no aprende.
  - Intermedio (0.01–0.03) ⇒ suele ser un buen compromiso: aprende a velocidad razonable y estable.
  - Muy alto (0.3, 1, 3…) ⇒ pasos enormes ⇒ el loss sube y baja, la frontera salta de un lado a otro ⇒ entrenamiento **inestable** o no converge.

- **Si cambio el batch size:**
  - Batch pequeño ⇒ más actualizaciones por época, gradiente más ruidoso pero a veces mejor generalización.
  - Batch grande ⇒ menos actualizaciones por época, gradiente más estable, pero puede atascarse en mínimos pobres.

- **Si activo/desactivo features (x1, x2, x1², x2², x1x2…):**
  - Quitar features ⇒ la red ve **menos información** ⇒ puede que el problema ya no sea separable.
  - Añadir features buenas (como x², x·y en XOR) ⇒ haces el problema **más fácil de separar** en el nuevo espacio.
  - Añadir features irrelevantes ⇒ metes ruido y a veces empeoras el aprendizaje.

- **Si aumento las épocas (Epochs):**
  - Pocas épocas ⇒ red no llega a una buena solución ⇒ *underfitting*.
  - Número adecuado ⇒ loss baja y se estabiliza en train y test.
  - Demasiadas épocas con modelo grande ⇒ test loss empieza a empeorar ⇒ *overfitting* (memoriza el entrenamiento).

Ten en mente siempre esta frase para responder oralmente:

> "Cuando aumento la capacidad del modelo (más capas/neurona o mejores features) la red puede aprender fronteras más complejas, pero si me paso aparece sobreajuste. El learning rate controla lo rápido y lo estable que se mueve la frontera, y las funciones de activación determinan qué tipo de no linealidades puede representar la red."

### 0.2. Chuleta de 2 horas (lee esto varias veces)

Si solo tienes un rato, memoriza estas ideas.

#### A. Qué significan los dibujos

- **Puntos**: azul/naranja = clase real.
- **Fondo**: color que predice la red.
- **Línea entre colores**: frontera de decisión.
- **Buen modelo**: la mayoría de puntos tienen el mismo color que el fondo.

#### B. Efecto de cada parámetro (una frase cada uno)

- **Más capas / más neuronas**: la frontera puede ser más curva y compleja ⇒ **más capacidad**, pero también **más overfitting** si me paso.
- **Menos capas / menos neuronas**: frontera muy simple ⇒ **underfitting** si el problema es complicado.
- **Learning rate muy bajo**: aprende **muy lento**, casi no se mueve la frontera.
- **Learning rate intermedio (0.01–0.03)**: suele dar un aprendizaje **rápido y estable**.
- **Learning rate muy alto**: la frontera salta mucho, el loss sube y baja ⇒ **entrenamiento inestable**.
- **Activación lineal**: casi solo fronteras rectas.
- **Tanh / ReLU**: permiten **curvas** y patrones no lineales (son las que quieres para Circle, XOR, etc.).
- **Activar x², x·y, etc.**: cambias el espacio de entrada ⇒ algunos problemas imposibles (XOR) se vuelven **separables**.
- **Épocas pocas**: no le dio tiempo a aprender.
- **Épocas demasiadas con modelo grande**: empieza a memorizar el entrenamiento ⇒ **overfitting**.

#### C. Cómo contestar "¿qué pasa si...?" en el examen

Siempre responde con esta estructura:

1. **Cambio**: "Si aumento (o reduzco) X…".
2. **Efecto visual**: "…la frontera se vuelve más (o menos) curva / detallada / estable…".
3. **Consecuencia**: "…lo que puede mejorar el ajuste, pero también aumenta el riesgo de underfitting/overfitting".

Ejemplos:

- "Si aumento el número de neuronas, la red tiene más capacidad para aprender patrones complejos y la frontera de decisión se vuelve más curva, pero si me paso aparece overfitting porque empieza a seguir el ruido de los datos." 
- "Si subo mucho el learning rate, los pasos de actualización son tan grandes que el entrenamiento se vuelve inestable: el loss oscila y la frontera no se estabiliza." 
- "Si activo las features cuadráticas en XOR, cambio el espacio de entrada y el problema deja de ser no linealmente separable, por eso la red por fin encuentra una frontera que separa bien las cuatro regiones."

### 0.3. Diccionario de la pantalla (nombre en inglés → significado en español)

Usa esto para asociar EXACTAMENTE lo que ves en la plataforma con el concepto.

#### Barra superior

- **Epoch**: número de épocas. 
  - Cada época = una pasada completa del dataset de entrenamiento.
- **Learning rate**: tasa de aprendizaje.
  - Qué tan grandes son los pasos al actualizar los pesos.
- **Activation**: función de activación de las capas ocultas.
  - Opciones típicas: ReLU, Tanh, Sigmoid, Linear.
- **Regularization**: tipo de regularización.
  - `None`, `L1`, `L2` (métodos para reducir overfitting penalizando pesos grandes).
- **Regularization rate**: fuerza de la regularización.
  - Cuánto penaliza pesos grandes (cuanto más alto, más fuerte la regularización).
- **Problem type**: tipo de problema.
  - `Classification` = clasificación (colores/clases).
  - `Regression` = regresión (valores continuos, curva).

#### Panel izquierdo – DATA

- **Which dataset do you want to use?**: qué conjunto de datos quieres usar.
  - Iconos:
    - Two clusters (dos nubes) → clasificación lineal.
    - Circle (círculo) → frontera circular.
    - XOR, espiral, etc.
- **Ratio of training to test data**: porcentaje de datos de entrenamiento frente a test.
  - Más entrenamiento = la red ve más ejemplos para aprender.
- **Noise**: nivel de ruido.
  - Cuanto más alto, más puntos mal colocados / mezclados.
- **Batch size**: tamaño del lote.
  - Número de ejemplos usados en cada actualización de pesos.
- **REGENERATE**: volver a generar el dataset (nuevos puntos).

#### Panel central – FEATURES y NETWORK

- **Which properties do you want to feed in?**: qué variables (features) quieres usar de entrada.
  - `X1`, `X2` → variables originales.
  - `X1²`, `X2²` → cuadrados (términos cuadráticos).
  - `X1·X2` → producto (interacción entre variables).
  - `sin(X1)`, `sin(X2)` → funciones seno de las variables.
- **+ / − (sobre las capas)**: añadir o quitar capas ocultas (`HIDDEN LAYER`).
- **1 HIDDEN LAYER**: hay una capa oculta.
- **1 neuron (debajo de la capa)**: número de neuronas en esa capa.

Colores y líneas:

- **Blue / Orange lines**: pesos entre neuronas.
  - Azul = peso positivo, naranja = peso negativo.
  - Grosor = magnitud del peso (qué tanto influye).
- **This is the output from one neuron. Hover to see it larger.**
  - Mapa de activación de esa neurona: te muestra qué zonas del plano activa o desactiva.

#### Panel derecho – OUTPUT

- **OUTPUT**: resultado de la red (predicciones sobre todo el plano).
- **Test loss**: pérdida (error) en datos de test.
- **Training loss**: pérdida (error) en datos de entrenamiento.
- **Show test data**: mostrar también los puntos de test.
- **Discretize output**: hace que el fondo sea solo dos colores sólidos (sin gradiente).
- Texto inferior: 
  - **"Colors shows data, neuron and weight values"**: los colores indican valores de datos, neuronas y pesos.

Con este diccionario puedes leer la pantalla en voz alta así: 

> "Aquí veo `Learning rate 0.003` (tasa de aprendizaje baja), `Activation ReLU` (función de activación no lineal rápida), `Problem type Classification` (estoy clasificando), una `1 hidden layer` con `1 neuron` (modelo muy simple) y el `Test loss 0.002` y `Training loss 0.003` me dicen que el error es muy bajo tanto en entrenamiento como en test."

### 0.4. Cómo saber si tu configuración es buena o mala (pasos claros)

Usa estos pasos SIEMPRE que hagas un ejercicio en el Playground.

#### Paso 1: Mira solo el dibujo (sin números)

- **Bueno**:
  - La mayoría de puntos azules están en zona de fondo azul.
  - La mayoría de puntos naranjas están en zona de fondo naranja.
  - La frontera (cambio de color) tiene una forma **parecida** al patrón de los datos (línea, círculo, XOR, etc.).
- **Malo**:
  - Muchos puntos de un color caen en fondo del color contrario.
  - La frontera no se parece al patrón que ves en los datos (por ejemplo, datos circulares y frontera casi recta).

#### Paso 2: Mira Training loss y Test loss

- **Training loss bajo y Test loss bajo** ⇒ **configuración buena** (aprende y generaliza bien).
- **Training loss alto y Test loss alto** ⇒ **underfitting** (modelo demasiado simple o pocas épocas).
- **Training loss muy bajo y Test loss alto** ⇒ **overfitting** (modelo demasiado complejo o demasiadas épocas / poco regularization).

#### Paso 3: Decide qué cambiar

- Si ves **underfitting** (no aprende):
  - Sube un poco la **capacidad**:
    - Más neuronas en la capa oculta.
    - O añade una segunda capa oculta.
  - Sube un poco el **número de épocas** (deja correr más tiempo).
  - Sube ligeramente el **learning rate** si está muy bajo.

- Si ves **overfitting** (entrena perfecto pero test mal, frontera muy retorcida):
  - Baja la **capacidad**:
    - Menos neuronas.
    - Menos capas.
  - Usa **Regularization = L2** con un `Regularization rate` pequeño (ej. 0.001).
  - Reduce un poco las **épocas** (para antes).

- Si ves **entrenamiento inestable** (loss sube y baja mucho, frontera salta):
  - Baja el **learning rate**.
  - Mantén igual las capas y neuronas para no mezclar efectos.

#### Paso 4: Escribe una frase de conclusión

Para cada ejercicio, intenta escribir una frase tipo:

> "Con esta configuración (`1 hidden layer`, `3 neurons`, `ReLU`, `learning rate 0.03`) la mayoría de los puntos quedan bien clasificados y los valores de `training loss` y `test loss` son bajos, por lo que el modelo tiene suficiente capacidad sin llegar a sobreajustar."

### 1. Ejercicio 1 – Clasificación lineal básica (Two clusters)

**Del PDF:** Problema 1 / Ejercicio 1.

**Objetivo:** Separar dos grupos de puntos bien separados por una **frontera casi lineal**.

#### 1.1. Configuración inicial recomendada

En `Data` selecciona el dataset **"Two clusters"**.

Configura los controles así (como dice el PDF):

- **Hidden layers**: 1
- **Neuronas en esa capa**: 1
- **Activation**: `Tanh`
- **Learning rate**: 0.03
- **Problem type**: `Classification`
- **Noise**: 0 (para empezar limpio)
- **Features activadas**: solo `x` y `y` (desactiva x², y², x·y, etc.)

Pulsa **Play ▶** para entrenar.

#### 1.2. Qué observar

Mientras avanza el entrenamiento mira:

- Cómo la **línea de decisión** (cambio de color del fondo) se va moviendo hasta separar bien los clusters.
- El **training loss** disminuyendo.
- Cuántas **épocas** tarda en estabilizarse.

#### 1.3. Experimento 1 – Cambiar función de activación

Vuelve a entrenar el mismo problema probando:

- `Tanh`
- `ReLU`
- `Sigmoid`

**Qué debes responder / entender:**

- ¿Con cuál la línea de decisión se ajusta más rápido?
- ¿Con cuál ves una convergencia más suave o más ruidosa?

**Explicación técnica para tu informe:**

- `Tanh` produce salidas entre -1 y 1, lo que puede ayudar cuando los datos están centrados.
- `ReLU` es muy utilizada en redes profundas, pero con tan pocas neuronas la diferencia puede ser menor.
- `Sigmoid` satura con facilidad, haciendo que el gradiente sea más pequeño y el aprendizaje más lento.

Puedes escribir algo así:

> "En un problema linealmente separable como Two Clusters, una red con una sola neurona oculta y función de activación tanh es suficiente. Al comparar tanh, ReLU y sigmoid, se observa que tanh y ReLU suelen converger más rápido, mientras que sigmoid presenta un aprendizaje más lento debido a la saturación de la activación (gradientes más pequeños)."

#### 1.4. Experimento 2 – Cambiar la tasa de aprendizaje

Usando la misma arquitectura (1 capa, 1 neurona, activation tanh):

Prueba con diferentes **learning rate**:

- `0.0001`
- `0.03` (base)
- `0.3`

**Qué observar:**

- Con `0.0001`: la línea se mueve muy despacio, el loss baja muy lentamente.
- Con `0.03`: aprendizaje razonablemente rápido y estable.
- Con `0.3`: la línea puede oscilar, el loss puede bajar y subir, o incluso no converger bien.

**Cómo explicarlo:**

> "La tasa de aprendizaje controla el tamaño de los pasos en el descenso de gradiente. Valores muy bajos producen un entrenamiento extremadamente lento, mientras que valores demasiado altos vuelven el proceso inestable y pueden impedir la convergencia. En este problema, un learning rate intermedio (≈0.03) logra un buen equilibrio entre velocidad y estabilidad."

### 3. Ejercicio 3 – Overfitting en clasificación (Gaussian)

**Del PDF:** Problema 3 / Ejercicio 3.

**Objetivo:** Observar cómo el aumento de complejidad del modelo puede llevar a **sobreajuste (overfitting)**.

#### 3.1. Configuración inicial (modelo simple)

En `Data`, selecciona el dataset **"Gaussian"**.

Configura la red como indica el PDF:

- **Hidden layers**: 1
- **Neuronas**: 2
- **Activation**: `Tanh` (o ReLU, según prefieras, pero mantén constante para comparar)
- **Learning rate**: 0.03 (o similar)
- **Features**: `x` y `y` activas.

Entrena hasta que el loss se estabilice.

**Qué verás:**

- Una frontera de decisión relativamente **suave**, que separa razonablemente los grupos pero sin volverse extremadamente irregular.

#### 3.2. Configuración compleja (posible overfitting)

Ahora aumenta la complejidad como sugiere el enunciado:

- **Hidden layers**: 3
- **Neuronas en cada capa**: 8

Es decir, una red de tipo: `input -> 8 -> 8 -> 8 -> output`.

Vuelve a entrenar.

**Qué deberías observar:**

- La frontera de decisión se vuelve **mucho más detallada**, con curvas y pliegues que se adaptan al contorno exacto de los puntos de entrenamiento.
- Puede que la clasificación del conjunto de entrenamiento mejore (menos puntos mal clasificados), pero la frontera se ve "demasiado ajustada" al patrón concreto.

#### 3.3. Cómo interpretar el overfitting

En términos técnicos:

- El modelo complejo tiene **muchos más parámetros** (pesos y biases), lo que le permite representar funciones muy complicadas.
- Esto reduce el error en entrenamiento, pero también hace que pueda **memorizar ruido** o detalles que no se generalizan bien.

Frase sugerida para tu informe:

> "Al incrementar la arquitectura de 1 capa oculta con 2 neuronas a 3 capas ocultas con 8 neuronas cada una, la frontera de decisión se vuelve mucho más irregular. Aunque el error en el conjunto de entrenamiento disminuye, la forma de la frontera indica sobreajuste, ya que el modelo comienza a ajustarse a detalles muy específicos de los datos en lugar de capturar la estructura general del problema."

También puedes mencionar la solución típica:

- Reducir complejidad (menos capas/neurona).
- Usar **regularización** (L2, dropout) o **early stopping**.
- Aumentar el tamaño del dataset si es posible.

### 4. Ejercicio 4 – Importancia de las variables de entrada (XOR)

**Del PDF:** Problema 4 / Ejercicio 4.

**Objetivo:** Ver cómo **agregar nuevas features derivadas** (x², y², x·y) ayuda a resolver un problema **no linealmente separable** como XOR.

#### 4.1. Configuración inicial sin features derivadas

En `Data`, selecciona el dataset **"XOR"**.

En la sección **Features**, haz lo siguiente:

- Activa solo:
  - `x1` (o `x`)
  - `x2` (o `y`)
- Desactiva:
  - `x1²`, `x2²`, `x1x2` (o equivalentes x², y², x·y)

Configura una red sencilla, por ejemplo:

- **Hidden layers**: 1
- **Neuronas**: 2 o 3
- **Activation**: `Tanh`
- **Learning rate**: 0.03

Entrena y observa.

**Qué deberías notar:**

- La red tiene dificultades para separar correctamente las cuatro regiones típicas del XOR.
- La frontera de decisión no consigue una separación limpia; quedan muchos puntos mal clasificados.

#### 4.2. Agregar características derivadas

Ahora, en **Features**, activa también:

- `x1²` (x²)
- `x2²` (y²)
- `x1x2` (x·y)

Deja la misma arquitectura de red y vuelve a entrenar.

**Qué cambia:**

- La red ahora puede encontrar una frontera de decisión que separa correctamente las cuatro regiones XOR.
- El fondo azul/naranja se ajusta a la forma lógica del patrón.

#### 4.3. Interpretación técnica

La idea clave es que estas nuevas variables derivadas **cambian el espacio de entrada**:

- En el espacio original (solo x1 y x2), el problema XOR **no es linealmente separable**.
- Al agregar x1², x2² y x1x2, estás mapeando los datos a un espacio de mayor dimensión donde **sí** es posible separarlos con una frontera más simple.

Frase para tu informe:

> "El problema XOR no es linealmente separable en el espacio original de dos variables (x1, x2). Al añadir características derivadas como x1², x2² y x1x2, el modelo dispone de un espacio de entrada más rico donde la frontera de decisión puede ser lineal o más sencilla. En el Playground se observa que, tras activar estas features adicionales, la red aprende a separar correctamente las cuatro regiones del problema XOR."

### 5. Problemas de clasificación compleja y ruido (Ejercicios 5 y similares)

**Del PDF:** Problema 5 y otros problemas de clasificación con patrones complejos o datos ruidosos.

**Objetivo:** Entender cómo influyen:

- El **tamaño del conjunto de datos**.
- El **número de capas ocultas**.
- El **número de neuronas**.

#### 5.1. Configurar un dataset complejo

En `Data`, puedes usar datasets como:

- **Spiral** (si aparece) o **moons**/patrones mezclados.
- O subir el nivel de **Noise** en datasets como Gaussian o Circle para simular datos ruidosos.

#### 5.2. Jugar con la complejidad del modelo

Prueba secuencialmente:

1. **Modelo simple**: 1 capa oculta con pocas neuronas (2–4).
2. **Modelo medio**: 2 capas ocultas con 4–6 neuronas cada una.
3. **Modelo grande**: 3–4 capas ocultas con 8 neuronas.

Mantén el **learning rate** razonable (0.03) para no mezclar efectos.

#### 5.3. Qué observar

- Cómo cambia la forma de la frontera de decisión:
  - Modelo simple → frontera suave, quizá no logra capturar todos los detalles.
  - Modelo grande → frontera muy compleja y retorcida, capaz de seguir el ruido.
- Cómo cambia el número de puntos mal clasificados en **train** y en **test** (si activas "Show test data").

#### 5.4. Interpretación general

- **Poca capacidad** (pocas capas/neurona) → *underfitting*: el modelo no tiene suficiente flexibilidad para capturar el patrón.
- **Demasiada capacidad** (muchas capas/neurona) → *overfitting*: el modelo memoriza detalles y ruido.

Frase tipo para el informe:

> "En datasets con patrones altamente complejos y presencia de ruido, aumentar la profundidad y el número de neuronas mejora inicialmente la capacidad del modelo para capturar la estructura subyacente. Sin embargo, más allá de cierto punto, la frontera de decisión se vuelve excesivamente irregular, indicando sobreajuste, ya que el modelo se adapta al ruido específico del conjunto de entrenamiento."

### 6. Problemas de regresión (Ejercicios 6, 7, 8, 9, 10)

En la parte de **regresión** del PDF se trabaja con relaciones entre variables continuas, no con clases.

En el Playground, cambia **Problem type** a `Regression`.

#### 6.1. Ejercicio 6 – Relación simple entre variables

- Elige un dataset de tipo **regresión simple** (por ejemplo, una recta con poco ruido).
- Usa una red muy sencilla:
  - Sin capas ocultas, o con 1 capa con pocas neuronas.

**Qué observar:**

- Una red sin capas ocultas se comporta como una **regresión lineal**.
- Con una capa oculta pequeña puedes aproximar relaciones algo más curvas.

Frase para el informe:

> "En problemas con una relación aproximadamente lineal entre variables, una red sin capas ocultas o con una única capa oculta pequeña es suficiente para aproximar la función objetivo. Modelos más complejos no aportan mejoras significativas y pueden incrementar el riesgo de sobreajuste." 

#### 6.2. Ejercicio 7 – Relación no lineal

- Usa un dataset con forma claramente no lineal (por ejemplo, una parábola o seno).
- Compara:
  - Modelo **sin** capas ocultas.
  - Modelo con **1 capa oculta** y varias neuronas (por ejemplo 4–6).

**Qué deberías ver:**

- El modelo sin capas ocultas no puede seguir bien la curva (underfitting).
- El modelo con capa oculta sí puede aproximar la forma no lineal.

#### 6.3. Ejercicio 8 – Influencia de la tasa de aprendizaje

- Mantén la misma arquitectura.
- Cambia solo el **learning rate** entre:
  - Muy bajo (0.0001)
  - Medio (0.01–0.03)
  - Alto (0.3 o más)

**Observa:**

- Con tasas muy bajas, la curva ajustada cambia muy despacio.
- Con tasas intermedias, converge de forma estable.
- Con tasas altas, el error puede oscilar mucho y la curva no se estabiliza.

Frase técnica:

> "En regresión, la tasa de aprendizaje determina la velocidad y estabilidad del descenso de gradiente. Un valor demasiado alto provoca oscilaciones en la función ajustada, mientras que un valor demasiado bajo ralentiza el entrenamiento sin mejorar necesariamente la calidad final del ajuste." 

#### 6.4. Ejercicio 9 – Complejidad del modelo en predicción

- Usa un dataset de regresión con cierta complejidad.
- Compara modelos con distinto número de capas y neuronas, igual que en clasificación.

**Conclusión general:**

- Más complejidad permite aproximar mejor la función en entrenamiento.
- Pero si la función subyacente no es tan compleja, un modelo grande sólo aprenderá el ruido → sobreajuste.

#### 6.5. Ejercicio 10 – Importancia de las variables en regresión

- Empieza con pocas features activadas.
- Añade después nuevas características derivadas (x², x·y, etc.).

**Qué observar:**

- Algunas nuevas variables **mejoran el ajuste** porque aportan información relevante (por ejemplo, permiten capturar curvatura).
- Otras pueden **no ayudar** o incluso perjudicar si sólo añaden ruido o redundancia.

Frase para el informe:

> "La selección de variables de entrada es crítica en problemas de regresión. Incluir características relevantes, como términos cuadráticos, permite que la red modele relaciones no lineales de forma más precisa. Sin embargo, añadir variables irrelevantes introduce ruido y puede dificultar el entrenamiento o llevar a modelos menos interpretables."

### 7. Checklist práctico para jugar en la plataforma y entender cada problema

Cada vez que resuelvas un ejercicio en `playground.tensorflow.org`, sigue este checklist:

1. **Identificar el tipo de problema**
   - ¿Es **clasificación** o **regresión**?
   - ¿La frontera esperada es lineal o claramente no lineal (círculo, XOR, espiral…)?

2. **Elegir una configuración inicial sencilla**
   - Pocas capas y pocas neuronas.
   - Learning rate intermedio (por ejemplo, 0.03).
   - Features mínimas necesarias (x, y).

3. **Entrenar y observar**
   - Forma de la frontera de decisión o de la curva de regresión.
   - Training loss (y, si lo muestras, test loss).
   - Número de épocas hasta que se estabiliza.

4. **Variar UN híperparámetro a la vez**
   - Número de neuronas/capas → capacidad / complejidad.
   - Función de activación → velocidad de aprendizaje y forma de la frontera.
   - Learning rate → velocidad y estabilidad del entrenamiento.
   - Features → qué tan separable es el problema en el nuevo espacio.

5. **Comparar y sacar conclusiones**
   - ¿El cambio mejoró o empeoró el resultado?
   - ¿Se ve underfitting (modelo demasiado simple) o overfitting (demasiado complejo)?
   - ¿Puedes justificarlo con conceptos como capacidad del modelo, no linealidad, tasa de aprendizaje, etc.?

6. **Escribir una mini conclusión por ejercicio**
   - 1–3 frases técnicas pero claras, del estilo de las que te he dejado en cada sección.
